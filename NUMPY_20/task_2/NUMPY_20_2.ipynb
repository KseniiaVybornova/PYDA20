Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')
```

```python
Baltic = data[ (data['country'] =='Latvia') | (data['country'] =='Lithuania') | (data['country'] =='Estonia') ]
```

```python
Baltic_category = Baltic[(Baltic['category'] == 4) | (Baltic['category'] == 12) | (Baltic['category'] == 21)]
```

```python
Baltic_period = Baltic_category[(Baltic_category['year'] >= 2005) & (Baltic_category['year'] <= 2010)]
```

```python
Baltic_period[Baltic_period['quantity'] > 0]['quantity'].sum()
```
